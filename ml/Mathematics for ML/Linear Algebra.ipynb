### Linear Algebra

* When formalising concepts, common approach is to construct a set of objects (symbols) and a set of rules to manipulate these objects. This is called as algebra
* Linear algebra is study of vectors and certain rules to manipulate them
* Vectors are objects that can be added together and multiplied by scalars to produce another object of same kind (any object that satisfies these two properties can be considered a vector)
* Examples of vectors
  - Geometric vectors are directed segments which can be drawn
      - Vectors $ \overrightarrow {\mathbf x} $ and $ \overrightarrow {\mathbf y} $ can be added such that $ \overrightarrow {\mathbf x} + \overrightarrow {\mathbf y} = \overrightarrow {\mathbf z}$ is another geometric vector
      - Multiplication by scalar $ \lambda \overrightarrow {\mathbf x}, \lambda \in \Re $ is also a geometric vector and is the original vector scaled by $ \lambda $
      - Interpreting vectors as geometric vectors enables us to use our intuitions about direction and magnitude to reason about mathematical operations
  - Polynomials are also vectors
      - Two polynomials can be added together, resulting in another polynomial, and they can be multipled by a scalar $\lambda \in \Re$ resulting in a polynomial
      - They are very different from geometric vectors. While geometric vectors are concrete drawings, polynomials are abstract concepts
  - Audio signals are vectors
      - Audio signals can be represented as a series of numbers, can be added and scaled
  - Elements of $ \Re^n $ (tuples of real numbers) are vectors. $\Re$ is more abstract than polynomials
  
     Example: $ a= 
    \begin{bmatrix}
    1 \\
    2 \\
    3 \\
    \end{bmatrix} \in \Re^3
    $
    
     Adding two vectors $ \mathbf a, \mathbf b \in \Re$ component wise results in another vector $\mathbf a + \mathbf b = \mathbf c \in \Re^n$. 
     Multiplying $\mathbf a \in \Re^n$ by $\lambda \in \Re$ results in a scaled vector $\lambda \mathbf a \in \Re^n$
 * We will largely focus on vectors in $\Re^n$ since most algorithms in linear algebra are formulated in $\Re^n$

* Span of vectors
    * The "span" of two vectors $ \overrightarrow {\mathbf v} $ and $ \overrightarrow {\mathbf w}$ is the set of all linear combinations of the two vectors $ a* \overrightarrow {\mathbf v} + b * \overrightarrow {\mathbf w}$
    * Span of most 2D vectors is the 2D space, as you can reach all the vectors by linear combination
    * Span of overlapping 2D vectors is a line
    * Span of two zero vectors is just the origin
    * Span of two vectors in 3D space would be a plane
    * Span of three vectors in 3D space woul be 3D space, if all three point in different directions
    * When we have multiple vectors and we can remove one without reducing the span (a vector is overlapping or already in span of other vectors), we call the vectors to be linearly dependent

#### Linear Transformations
  * When applying linear transformation to vectors, we only need to record transformed vectors of $ \hat{i} $ and $ \hat{j} $. The new vector can be deduced by multiplying scalars with new location of $\hat {i}$ and $\hat{j}$
  * Ex: $\overrightarrow {\mathbf v} = -1 \hat{i} + 2 \hat{j}$.
    Say transformed $\hat{i}=\begin{bmatrix}
    1 \\
    -2 \\
    \end{bmatrix}$ and transformed $\hat{j}=\begin{bmatrix}
    3 \\
    0 \\
    \end{bmatrix}$, then the new vector is 
    $-1 * \begin{bmatrix}
    1 \\
    -2 \\
    \end{bmatrix} + 2 * \begin{bmatrix}
    3 \\
    0 \\
    \end{bmatrix} = \begin{bmatrix}
    5 \\
    2 \\
    \end{bmatrix}$
    
* Matrices as Linear Transformations
  * A 2X2 matrix can be viewed as two column vectors representing transformed $\hat{i}$ and transformed $\hat{j}$
  * For example, the matrix $\begin{bmatrix}
    3 & 2\\
    -2 & 1\\
    \end{bmatrix}$ can be interpreted as two column vectors: transformed $\hat{i}=\begin{bmatrix} 3 \\ -2 \end{bmatrix}$ and transformed $\hat{j}=\begin{bmatrix} 2 \\ 1 \end{bmatrix}$
  * Therefore, if we have any matrix and a vector, we can compute the transformed co-ordinates of the vector by multiplying it with the matrix
  * Example, the vector $\begin{bmatrix} 5 \\ 7\end{bmatrix}$ will be transformed with the above matrix as 
    $ 5 * \begin{bmatrix} 3 \\ -2 \end{bmatrix} + 7 * \begin{bmatrix} 2 \\ 1\end{bmatrix}$
  * In the general case, when matrix is $\begin{bmatrix} a & b \\ c & d \end{bmatrix}$, $\begin{bmatrix} a \\ c \end{bmatrix} $ is where the first basis vector lands after transformation, and $\begin{bmatrix} b \\ d\end{bmatrix}$ is where the second basis vector lands. 
  * When this transformation is applied to any vector $\begin{bmatrix}x \\ y \end{bmatrix}$, the final co-ordinates will be $ x * \begin{bmatrix} a \\ c \end{bmatrix} + y * \begin{bmatrix} b \\ d\end{bmatrix}$ = $\begin{bmatrix} a*x + b*y \\ c*x + d*y \end{bmatrix}$. 
  * This is nothing but the matrix-vector multiplication $\begin{bmatrix} a & b \\ c & d \end{bmatrix}$ * $\begin{bmatrix}x \\ y \end{bmatrix}$
  * Few examples of commonly used transformations:
      * 90 degree rotation: If we rotate of space by 90 degrees counterclock wise, transformed $\hat{i}=\begin{bmatrix} 0 \\ 1 \end{bmatrix}$ and transformed $\hat{j}=\begin{bmatrix} -1 \\ 0 \end{bmatrix}$. Therefore the matrix for 90 degree counterclock wise rotation is $\begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix}$. So to get the transformed co-ordinates for any vector $\begin{bmatrix} x \\ y \end{bmatrix}$, we simply need to multiply it with the transformation matrix.
      * Similarly, shear transformation has $\hat{i}$ fixed, but $\hat{j}$ transforms to $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$. Therefore, the shear matrix is $\begin{bmatrix} 1 & 1 \\ 0 & 1\end{bmatrix}$
      
* Matrix Multiplication as Composition
  * Say we apply two transformations one after the other: rotation and then shear. Then the final matrix can be represented by following where the $\hat{i}$ and $\hat{j}$ land. Therefore the matrix $\begin{bmatrix} 1 & -1 \\ 1 & 0 \end{bmatrix}$ (obtained by visually tracking the co-ordinates) has the same effect as applying two sequential transformations of rotation and shear.
  * We can see that $\begin{bmatrix} 1 & 1 \\ 0 & 1\end{bmatrix}$ * $\begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix}$ also equals to the same final matrix. 
  * This can be seen as:
  ($\begin{bmatrix} 1 & 1 \\ 0 & 1\end{bmatrix}$) * ($\begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix}$) * ($\begin{bmatrix} x \\ y\end{bmatrix}$) = ($\begin{bmatrix} 1 & 1 \\ 0 & 1\end{bmatrix}$ * $\begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix}$) * $\begin{bmatrix} x \\ y\end{bmatrix}$ = $\begin{bmatrix} 1 & -1 \\ 1 & 0 \end{bmatrix}$ * $\begin{bmatrix} x \\ y\end{bmatrix}$

#### Systems of Linear Equations

* In general for a real-valued system of linear equations, we obtain no, exactly one or infinitely many solutions
* In a system of equations with two variables $x_1$ and $x_2$, each linear equation defines a line on the $x_1, x_2$ plane
* Since the solution to system of linear equations must satisfy all equations simultaneously, the solution set is the intersection of these lines
* This intersection set can be a line (if the equations describe the same line), a point, or empty (when lines are parallel)
* Similarly for three variables, each linear equation determines a plane in the 3D space. When these planes intersect (satisfy all linear equations), we obtain a solution set that can either be a plane, a line, a point or empty (when planes have no intersection). For visualisation, check [this video](https://www.youtube.com/watch?v=2q0HMHY-2iM)

```python
#Plot intersection of 2D equations: 4x1+4x2=5; 2x1-4x2=1
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# line1: when x2=0, x1=1.25; when x1=0, x2=1.25
ax.axline((1.25,0),(0,1.25))

# line2: when x2=0, x1=0.5; when x1=0, x2=-0.25
ax.axline((0.5,0),(0,-0.25))

ax.grid(True, which='both')
plt.show()
```

```python
import numpy as np
# Plot intersections of 3D equations: 

# create the figure
fig = plt.figure()

# add axes
ax = fig.add_subplot(111,projection='3d')

xx, yy = np.meshgrid(range(10), range(10))
z = (9 - 3*xx - yy) / 2 
z1= (xx - 5* yy) 

# plot the plane
ax.plot_surface(xx, yy, z, alpha=0.5)
ax.plot_surface(xx, yy, z1, alpha=0.5)

plt.show()
```
